```python
X = pd.DataFrame(pd.read_csv('/kaggle/input/2021-ml-p2/x_train.csv',engine='python',encoding='CP949'))
y = pd.DataFrame(pd.read_csv('/kaggle/input/2021-ml-p2/y_train.csv',engine='python',encoding='CP949'))
test = pd.DataFrame(pd.read_csv('/kaggle/input/2021-ml-p2/x_test.csv',engine='python',encoding='CP949'))
sample = pd.DataFrame(pd.read_csv('/kaggle/input/2021-ml-p2/sample_submission.csv',engine='python',encoding='CP949'))
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split #train set과 test set으로 데이터 나누기 위한 함수 호출
X_train,X_test,y_train,y_test=train_test_split(X,y, test_size=0.33, random_state=0)
X_train.shape
```

```python
# 표준화
from sklearn.preprocessing import StandardScaler, MinMaxScaler
sc = StandardScaler()
sc.fit(X_train)
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)

# 표준화
sc = MinMaxScaler()
sc.fit(X_train)
X_train_m = sc.transform(X_train)
X_test_m = sc.transform(X_test)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR
param_grid={
    'C':[1e3,5e3,1e4],
    'gamma':[0.001,0.005,0.01,0.1],
}
clf=GridSearchCV(SVR(kernel = 'rbf'), param_grid, cv=3)   
clf.fit(X,y)
```

```python
print(clf.best_params_)
```

```python
from sklearn.svm import SVR
clf = SVR(C = 3000, gamma = 0.25, kernel = 'rbf')
clf.fit(X_train_std, y_train)
print(clf.score(X_test_std, y_test))
```

```python
from sklearn.svm import SVR
clf = SVR(C = 5000, gamma = 0.25, kernel = 'rbf')
clf.fit(X_train_std, y_train)
print(clf.score(X_test_std, y_test))
```

```python
from sklearn.svm import SVR
clf = SVR(C = 7000, gamma = 0.25, kernel = 'rbf')
clf.fit(X_train_std, y_train)
print(clf.score(X_test_std, y_test))
```

```python
from sklearn.svm import SVR
sc = StandardScaler()
sc.fit(X)
X_std = sc.transform(X)
x_test_std= sc.transform(test)
clf = SVR(C = 5000, gamma = 0.25, kernel = 'rbf')

clf.fit(X_std, y)

x_test_pred = clf.predict(x_test_std)
```

```python
sample['predict'] = x_test_pred
sample.head()
sample.to_csv("./result-svm.csv", mode='w', index=False) 
```
